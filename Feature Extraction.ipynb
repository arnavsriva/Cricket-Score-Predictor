```python
import pandas as pd
import pickle
```

```python
data = pickle.load(open('data_v2.pkl','rb'))
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['city'].isnull()
```

```python
import numpy as np
cities = np.where(data['city'].isnull(), data['venue'].str.split().apply(lambda x: x[0]),data['city'])
```

```python
data['city'] = cities
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['venue'], inplace= True)
```

```python
eligible_cities = data['city'].value_counts()[data['city'].value_counts()>600].index.tolist()
```

```python
data = data[data['city'].isin(eligible_cities)]
```

```python
data
```

```python
data['current_score'] = data.groupby('match_id').cumsum()['runs']
```

```python
data['over'] = data['ball'].apply(lambda x:str(x).split(".")[0])
data['ball_no'] = data['ball'].apply(lambda x:str(x).split(".")[1])
data
```

```python
data['balls_bowled'] = (data['over'].astype('int')*6) + data['ball_no'].astype('int')
data
```

```python
data['balls_left'] = 120 - data['balls_bowled']
data['balls_left'] = data['balls_left'].apply(lambda x:0 if x<0 else x)
data
```

```python
data['player_dismissed'] = data['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
data['player_dismissed'] = data['player_dismissed'].astype('int')
data['player_dismissed'] = data.groupby('match_id').cumsum()['player_dismissed']
data['wickets_left'] = 10 - data['player_dismissed']
```

```python
data['crr'] = (data['current_score']*6)/data['balls_bowled']
```

```python
groups = data.groupby('match_id')

match_ids = data['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
data['last_five'] = last_five
```

```python
final_df = data.groupby('match_id').sum()['runs'].reset_index().merge(data,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
